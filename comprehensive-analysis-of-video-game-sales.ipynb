# **2.Collecting the data:**

**About Dataset:**
The dataset used for this analysis comprises the following columns, each providing key information about video games and their sales:

1. **Rank**: The numerical ranking or position of the video game in terms of sales or popularity.
2. **Name**: The title or name of the video game.
3. **Platform**: The gaming platform or console on which the game is released (e.g., PlayStation, Xbox, PC).
4. **Year**: The year when the game was released.
5. **Genre**: The category or genre of the video game (e.g., action, sports, role-playing).
6. **Publisher**: The company responsible for publishing and distributing the game.
7. **NA_Sales**: Sales in North America, typically measured in millions of units.
8. **EU_Sales**: Sales in Europe, typically measured in millions of units.
9. **JP_Sales**: Sales in Japan, typically measured in millions of units.
10. **Other_Sales**: Sales in other regions or areas, typically measured in millions of units.
11. **Global_Sales**: The total worldwide sales of the game, which is the sum of sales across all regions.

These columns collectively provide a comprehensive overview of each video game's performance in terms of sales, popularity, genre, and geographic distribution. The dataset's rich information enables analyses to uncover trends, preferences, and success factors within the video game industry, supporting a wide range of insights and decision-making processes for professionals and enthusiasts in the gaming domain.

**Correlation Analysis:**

The correlation coefficient ranges from -1 to 1:

* **Positive Correlation (0 to 1)**: As one variable increases, the other tends to also increase. A correlation coefficient closer to 1 indicates a stronger positive correlation.

* **Negative Correlation (-1 to 0)**: As one variable increases, the other tends to decrease. A correlation coefficient closer to -1 indicates a stronger negative correlation.

* **No Correlation (0)**: There is no linear relationship between the two variables.

```python
import seaborn as sns
# Selecting only numeric columns for correlation analysis
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns


correlation_matrix = df[numeric_columns].corr() # Calculating the correlation matrix for numeric columns

# Visualization
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap='viridis', annot=True, fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

* A correlation coefficient of -0.07 between 'Global Sales' and 'Year' suggests a very weak negative correlation between these two variables. The negative sign indicates that, on average, as the 'Year' increases, 'Global Sales' tends to slightly decrease. However, the small magnitude of the correlation coefficient suggests that the relationship is quite weak.

* A correlation coefficient of 0.94 between 'NA Sales' and 'Global Sales' indicates a very strong positive correlation between these two variables.

```python
import matplotlib.pyplot as plt

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['na_sales'], df['global_sales'], alpha=0.5)
plt.title('Scatter Plot: NA Sales vs Global Sales')
plt.xlabel('NA Sales (millions)')
plt.ylabel('Global Sales (millions)')
plt.grid(True)
plt.show()
```

# **1. Top Sellers**
# Which video games hold the top positions in terms of global sales?

```python
top_sellers = df[['name', 'global_sales']].nlargest(10, 'global_sales')

# Set the 'name' column as the index
top_sellers.set_index('name', inplace=True)

# Plotting
plt.figure(figsize=(7, 4))
top_sellers.plot(kind='bar', color='blue')
plt.title('Top Selling Games')
plt.xlabel('Game Name')
plt.ylabel('Global Sales (millions)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# **2. Platform Popularity:**
# What is the most preferred gaming platform based on sales?

```python
platform_popularity = df.groupby('platform')['global_sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
platform_popularity.plot(kind='bar', color='skyblue')
plt.title('Platform Popularity Based on Global Sales')
plt.xlabel('Platform')
plt.ylabel('Global Sales (millions)')
plt.show()
```

# **3. Platform with Highest Genre Sales**
# Which genre tends to have the highest sales globally?

```python
genre_success = df.groupby('genre')['global_sales'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
genre_success.plot(kind='bar', color='green')
plt.title('Genre Success Based on Global Sales')
plt.xlabel('Genre')
plt.ylabel('Average Global Sales (millions)')
plt.show()
```

# **4.Publisher Impact:**
#   Which publishers have the most top-ranking games?

```python
top_publishers = df.groupby('publisher')['name'].count().nlargest(10)
print("Top Publishers:")
print(top_publishers)
```

# 5. **Temporal Trends:**
#     How have video game sales evolved over the years?

```python

yearly_sales = df.groupby('year')['global_sales'].sum()
max_sales_year = yearly_sales.idxmax()

# Visualization with a marker for the year with maximum sales
plt.figure(figsize=(12, 6))
yearly_sales.plot(kind='line', marker='o', color='purple',label='total sales')
plt.axvline(x=max_sales_year, color='red', linestyle='--', label=f'Max Sales Year: {max_sales_year}')
plt.title('Total Global Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Sales (millions)')
plt.legend()
plt.show()
```

# 6. **Regional Preferences:** 
#    What are the most popular genres in North America, Europe, and Japan?

```python
df_region_sales = df[['genre', 'na_sales', 'eu_sales', 'jp_sales']].dropna()
total_sales_by_genre = df_region_sales.groupby('genre').sum()# Group by genre and calculate total sales for each region
plt.figure(figsize=(14, 8))# Plotting for North America, Europe, and Japan in a single graph
total_sales_by_genre['na_sales'].sort_values(ascending=False).plot(kind='bar', color='skyblue', position=0, width=0.25)# Plotting for North America
total_sales_by_genre['eu_sales'].sort_values(ascending=False).plot(kind='bar', color='lightgreen', position=1, width=0.25)# Plotting for Europe
total_sales_by_genre['jp_sales'].sort_values(ascending=False).plot(kind='bar', color='pink', position=2, width=0.25)# Plotting for Japan
plt.title('Top Genres in Different Regions')# Set labels and title
plt.xlabel('Genre')
plt.ylabel('Total Sales (millions)')
plt.legend(['North America', 'Europe', 'Japan']) # Set legend
plt.show() # Show the plot
```

#  **Annual global sales by region over time**

```python
#Annual global sales by region over time

df_region_sales = df[['year', 'na_sales', 'eu_sales', 'jp_sales', 'other_sales']].dropna()
annual_global_sales = df_region_sales.groupby('year').sum()# Group by year and calculate annual global sales for each region
plt.figure(figsize=(12, 6))                                # Plotting a line graph for annual global sales by region
plt.plot(annual_global_sales.index, annual_global_sales['na_sales'], label='North America', marker='o') # Plot lines for each region
plt.plot(annual_global_sales.index, annual_global_sales['eu_sales'], label='Europe', marker='o')
plt.plot(annual_global_sales.index, annual_global_sales['jp_sales'], label='Japan', marker='o')
plt.plot(annual_global_sales.index, annual_global_sales['other_sales'], label='Other Regions', marker='o')
plt.title('Annual Global Sales by Region Over Time')# Set title, xlabel, and ylabel
plt.xlabel('Year')
plt.ylabel('Global Sales (millions)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1)) # Show legend
plt.show()# Show the plot
```
